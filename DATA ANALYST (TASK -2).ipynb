# HR DATA ANALYSIS

```python
import pandas as pd

df= pd.read_csv('HR Data.csv').fillna(0)
dataframe = pd.DataFrame(df)
print(dataframe)
```

```python
shape = dataframe.shape
print('SHAPE: ', shape)
```

```python
dataframe.describe()
```

```python
null = dataframe.isna().sum()
print('NULL VALUES COUNT\n',null)
```

```python
# droping columns

df.drop(df.iloc[:, [8,9,10,12,13,14,15,16,17,19,20,21,25,27,28,29,30,31,32,33,34]], inplace=True, axis=1)
df
```

```python
# renaming columns

df = df.rename(columns = {'BusinessTravel': 'Business_Travel', 'DailyRate': 'Daily_Rate', 'DistanceFromHome': 'Distance_From_Home', 'EducationField': 'Education_Field', 'MonthlyIncome': 'Monthly_Income', 'OverTime': 'Over_Time', 'PercentSalaryHike': 'Percent_Salary_Hike', 'PerformanceRating': 'Performance_Rating', 'StandardHours': 'Standard_Hours'})
df.head(10)
```

```python
# droping duplicates

df = df.drop_duplicates()
df.head(5)
```

```python
# droping nan values

df = df.dropna()
df.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
df['Attrition'].value_counts().plot(kind='bar', title = 'ATTRITION')
plt.show()
```

```python
plt.figure(figsize=(10,10))
df['Over_Time'].value_counts().plot(kind='bar', title = 'OVER TIME')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.scatterplot(x="Age", y="Monthly_Income", hue="Gender", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['Performance_Rating'].plot(kind='kde', title = 'PEFORMANCE RATING')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title('ATTRITION - DEPARTMENT')
plt.plot()
```
